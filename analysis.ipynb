```python

import pandas as pd
from datetime import datetime

# Sample DataFrame (assuming df1 is your DataFrame with a Birthday column)
# df1 = pd.DataFrame({
#     'Birthday': ['1990-01-01', '1985-05-15', '2000-07-20', '1975-12-30', '1939-07-03']
# })

# Convert the Birthday column to datetime
df1['Birthday'] = pd.to_datetime(df1['Birthday'])

# Calculate Age
today = datetime.today()
df1['Age'] = today.year - df1['Birthday'].dt.year

# Adjust for whether the birthday has occurred this year
df1['Age'] -= ((df1['Birthday'].dt.month > today.month) | 
                ((df1['Birthday'].dt.month == today.month) & (df1['Birthday'].dt.day > today.day))).astype(int)

# Format the Birthday as "DD-MM-YYYY"
df1['Birthday'] = df1['Birthday'].dt.strftime('%d-%m-%Y')

# Display the updated DataFrame with Birthday and Age
print(df1[['Birthday', 'Age']].to_string(index=False))
```

```python
import pandas as pd
from datetime import datetime

# Sample DataFrame (assuming df1 is your DataFrame with a Birthday column)
# df1 = pd.DataFrame({
#     'Birthday': ['1990-01-01', '1985-05-15', '2000-07-20', '1975-12-30', '1939-07-03']
# })

# Convert the Birthday column to datetime
df1['Birthday'] = pd.to_datetime(df1['Birthday'])

# Calculate Age
today = datetime.today()
df1['Age'] = today.year - df1['Birthday'].dt.year

# Adjust for whether the birthday has occurred this year
df1['Age'] -= ((df1['Birthday'].dt.month > today.month) | 
                ((df1['Birthday'].dt.month == today.month) & (df1['Birthday'].dt.day > today.day))).astype(int)

# Define Age Groups
bins = [-1, 24, 40, float('inf')]  # Define the bin edges
labels = ['Youth', 'Young Adult', 'Adult']  # Define the corresponding labels

# Create Age Group column using pd.cut
df1['Age Group'] = pd.cut(df1['Age'], bins=bins, labels=labels)

# Format the Birthday as "DD-MM-YYYY"
df1['Birthday'] = df1['Birthday'].dt.strftime('%d-%m-%Y')

# Display the updated DataFrame with Birthday, Age, and Age Group
print(df1[['Birthday', 'Age', 'Age Group']].to_string(index=False))
```

# Distribution of customers by AGE Groups

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Count of customers in each Age Group
age_group_counts = df1['Age Group'].value_counts().sort_index()

# Step 2: Display the counts
print("Distribution of Customers by Age Group:")
print(age_group_counts)

# Calculate the total count for percentage calculations
total_count = age_group_counts.sum()

# Step 3: Create a donut chart
plt.figure(figsize=(10, 6))

# Function to display both count and percentage on the donut chart
def label_with_count_and_percentage(pct, allvalues):
    absolute = int(round(pct / 100 * total_count))  # Calculate the absolute count
    return f'{absolute} ({pct:.1f}%)'  # Return formatted string

# Create a pie chart with the specified colors
plt.pie(age_group_counts, labels=age_group_counts.index, autopct=lambda pct: label_with_count_and_percentage(pct, age_group_counts), 
        startangle=140, colors=sns.color_palette('Set2'))

# Draw a circle in the center of the pie to make it a donut chart
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.title('Distribution of Customers by Age Groups')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is circular.
plt.legend(age_group_counts.index, title="Age Groups", loc="upper right")  # Add a legend
plt.show()
```

# Distribution of customers by Gender

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Count of customers in each Gender
gender_counts = df1['Gender'].value_counts().sort_index()

# Step 2: Display the counts
print("Distribution of Customers by Gender:")
print(gender_counts)

# Calculate percentages
gender_percentages = (gender_counts / gender_counts.sum()) * 100

# Step 3: Plot the Gender Distribution Bar Chart
plt.figure(figsize=(10, 6))
bar_plot = sns.barplot(x=gender_counts.index, y=gender_counts.values, palette='Set2')

# Customize the chart
plt.title('Distribution of Customers by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)

# Add data labels with percentages on top of the bars
for index, value in enumerate(gender_counts.values):
    percentage = gender_percentages[index]
    plt.text(index, value, f'{value} ({percentage:.1f}%)', ha='center', va='bottom')

plt.legend(title="Gender", labels=gender_counts.index)  # Add a legend
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set the color palette
palette = sns.color_palette('Set3', 10)  # You can choose other palettes like 'Set2', 'husl', etc.

# Step 1: Count of customers in the top 10 cities
top_cities = df1['City'].value_counts().nlargest(10)

# Display the counts for top 10 cities
print("Top 10 Cities Distribution of Customers:")
print(top_cities)

# Calculate the total count for percentage calculations
total_city_count = top_cities.sum()

# Step 2: Prepare data for the top 10 cities stacked bar chart
city_data = pd.DataFrame({
    'Count': top_cities,
    'Percentage': (top_cities / total_city_count) * 100
})

# Step 3: Plot the Top 10 Cities Bar Chart
plt.figure(figsize=(10, 6))
plt.bar(city_data.index, city_data['Count'], label='Count', color=palette)

# Plot percentage on top of the count bars
for index, value in enumerate(city_data['Count']):
    plt.text(index, value, f'{value} ({city_data["Percentage"].iloc[index]:.1f}%)', ha='center', va='bottom')

# Adding labels and title
plt.title('Top 10 Cities Distribution of Customers')
plt.xlabel('City')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)
plt.legend(title='Counts')

# Display the plot
plt.tight_layout()
plt.show()

# Step 4: Count of customers in the top 10 states
top_states = df1['State'].value_counts().nlargest(10)

# Display the counts for top 10 states
print("Top 10 States Distribution of Customers:")
print(top_states)

# Calculate the total count for percentage calculations
total_state_count = top_states.sum()

# Step 5: Prepare data for the top 10 states stacked bar chart
state_data = pd.DataFrame({
    'Count': top_states,
    'Percentage': (top_states / total_state_count) * 100
})

# Step 6: Plot the Top 10 States Bar Chart
plt.figure(figsize=(10, 6))
plt.bar(state_data.index, state_data['Count'], label='Count', color=palette)

# Plot percentage on top of the count bars
for index, value in enumerate(state_data['Count']):
    plt.text(index, value, f'{value} ({state_data["Percentage"].iloc[index]:.1f}%)', ha='center', va='bottom')

# Adding labels and title
plt.title('Top 10 States Distribution of Customers')
plt.xlabel('State')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)
plt.legend(title='Counts')

# Display the plot
plt.tight_layout()
plt.show()

# Step 7: Count of customers in the top 10 countries
country_counts = df1['Country'].value_counts().nlargest(10)

# Display the counts for top 10 countries
print("Distribution of Customers by Countries:")
print(country_counts)

# Calculate the total count for percentage calculations
total_country_count = country_counts.sum()

# Step 8: Prepare data for the top 10 countries stacked bar chart
country_data = pd.DataFrame({
    'Count': country_counts,
    'Percentage': (country_counts / total_country_count) * 100
})

# Step 9: Plot the Top 10 Countries Bar Chart
plt.figure(figsize=(10, 6))
plt.bar(country_data.index, country_data['Count'], label='Count', color=palette)

# Plot percentage on top of the count bars
for index, value in enumerate(country_data['Count']):
    plt.text(index, value, f'{value} ({country_data["Percentage"].iloc[index]:.1f}%)', ha='center', va='bottom')

# Adding labels and title
plt.title('Distribution of Customers by Countries')
plt.xlabel('Country')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)
plt.legend(title='Counts')

# Display the plot
plt.tight_layout()
plt.show()

# Step 10: Count of customers in the top 10 continents
continent_counts = df1['Continent'].value_counts().nlargest(10)

# Display the counts for top 10 continents
print("Distribution of Customers by Continents:")
print(continent_counts)

# Calculate the total count for percentage calculations
total_continent_count = continent_counts.sum()

# Step 11: Prepare data for the top 10 continents stacked bar chart
continent_data = pd.DataFrame({
    'Count': continent_counts,
    'Percentage': (continent_counts / total_continent_count) * 100
})

# Step 12: Plot the Top 10 Continents Bar Chart
plt.figure(figsize=(10, 6))
plt.bar(continent_data.index, continent_data['Count'], label='Count', color=palette)

# Plot percentage on top of the count bars
for index, value in enumerate(continent_data['Count']):
    plt.text(index, value, f'{value} ({continent_data["Percentage"].iloc[index]:.1f}%)', ha='center', va='bottom')

# Adding labels and title
plt.title('Distribution of Customers by Continents')
plt.xlabel('Continent')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)
plt.legend(title='Counts')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set the color palette
palette = sns.color_palette('Set3', 10)

def plot_top_distribution(data, title, xlabel):
    """Function to plot top 10 distribution."""
    top_counts = data.value_counts().nlargest(10)
    total_count = top_counts.sum()

    # Prepare data for plotting
    plot_data = pd.DataFrame({
        'Count': top_counts,
        'Percentage': (top_counts / total_count) * 100
    })

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.bar(plot_data.index, plot_data['Count'], color=palette)
    for index, value in enumerate(plot_data['Count']):
        plt.text(index, value, f'{value} ({plot_data["Percentage"].iloc[index]:.1f}%)', ha='center', va='bottom')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel('Number of Customers')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    print(f"{title}:")
    print(top_counts)  # Display counts

# Step 1: Plot for Top 10 Cities
plot_top_distribution(df1['City'], 'Top 10 Cities Distribution of Customers', 'City')

# Step 2: Plot for Top 10 States
plot_top_distribution(df1['State'], 'Top 10 States Distribution of Customers', 'State')

# Step 3: Plot for Top 10 Countries
plot_top_distribution(df1['Country'], 'Top 10 Countries Distribution of Customers', 'Country')

# Step 4: Plot for Top 10 Continents
plot_top_distribution(df1['Continent'], 'Top 10 Continents Distribution of Customers', 'Continent')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Assuming df2, df3, and df5 are already loaded as DataFrames

# Step 1: Merge df2 (Sales) with df3 (Products) and df5 (Exchanges)
merged_sales_products = pd.merge(df2, df3, on='ProductKey', how='left')
merged_sales_exchanges = pd.merge(
    merged_sales_products, df5, 
    left_on=['Currency Code', 'Order Date'], 
    right_on=['Currency Code', 'Date'], 
    how='left'
)

# Calculate sales in original currency and in USD
merged_sales_exchanges['Total_Sales_Original'] = merged_sales_exchanges['Quantity'] * merged_sales_exchanges['Unit Price USD']
merged_sales_exchanges['Total_Sales_USD'] = merged_sales_exchanges['Total_Sales_Original'] * merged_sales_exchanges['Exchange']

# Group by Currency_Code to get total sales in both original currency and USD
currency_sales = merged_sales_exchanges.groupby('Currency Code').agg({
    'Total_Sales_Original': 'sum',
    'Total_Sales_USD': 'sum'
})

# Calculate impact (difference and percentage change)
currency_sales['Exchange_Rate_Impact'] = currency_sales['Total_Sales_USD'] - currency_sales['Total_Sales_Original']
currency_sales['Exchange_Rate_Percentage_Impact'] = ((currency_sales['Total_Sales_USD'] - currency_sales['Total_Sales_Original']) / currency_sales['Total_Sales_Original']) * 100

# Step 2: Plot Original vs. USD Sales by Currency Code
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Plot Original vs. USD Sales
currency_sales[['Total_Sales_Original', 'Total_Sales_USD']].plot(kind='bar', ax=ax[0], color=['skyblue', 'salmon'])
ax[0].set_title('Total Sales by Currency (Original vs. USD)')
ax[0].set_xlabel('Currency Code')
ax[0].set_ylabel('Total Sales')
ax[0].legend(['Original Currency', 'USD'], loc='upper left')
ax[0].set_xticklabels(currency_sales.index, rotation=45)

# Add value labels
for index, (original, usd) in enumerate(zip(currency_sales['Total_Sales_Original'], currency_sales['Total_Sales_USD'])):
    ax[0].text(index - 0.15, original, f'{original:,.0f}', ha='center', va='bottom', fontsize=9, color='blue')
    ax[0].text(index + 0.15, usd, f'{usd:,.0f}', ha='center', va='bottom', fontsize=9, color='red')

# Step 3: Plot Exchange Rate Impact as Percentage Change
currency_sales['Exchange_Rate_Percentage_Impact'].plot(kind='line', marker='o', ax=ax[1], color='purple')
ax[1].set_title('Exchange Rate Impact on Sales (%)')
ax[1].set_xlabel('Currency Code')
ax[1].set_ylabel('Impact (%)')
ax[1].set_xticks(range(len(currency_sales)))
ax[1].set_xticklabels(currency_sales.index, rotation=45)

# Add percentage labels
for index, percentage in enumerate(currency_sales['Exchange_Rate_Percentage_Impact']):
    ax[1].text(index, percentage, f'{percentage:.2f}%', ha='center', va='bottom', fontsize=9, color='purple')

plt.tight_layout()
plt.show()
```
